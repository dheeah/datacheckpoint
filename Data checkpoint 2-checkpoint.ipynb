```python
import requests

download_url = "https://raw.githubusercontent.com/dheeah/dataset/main/titanic-passengers.csv"
target_csv_path = "titanic_passengers.csv"

response = requests.get(download_url)
response.raise_for_status()    
with open(target_csv_path, "wb") as f:
    f.write(response.content)
print("Download ready.")
```
